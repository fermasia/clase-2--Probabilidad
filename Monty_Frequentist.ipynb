# Define Experiment

```python
def monty_iter():
    '''
    Monty Python Simulation
    Returns a list with 3 positions that can take 1 or 0 value
    1 = hit / 0 = miss
    1st position: result if sticking to first choice
    2nd position: result if changing to the option
    3rt position: result if sticking or changing is chosen randomly
    '''
    import random
    
    #set of options
    options = [1,2,3]
    
    # simulate a winning position
    winner = random.choice(options)
    
    primera_opcion = random.choice(options)
    
    descarte = [x for x in options if x not in [winner,primera_opcion]]
    if len(descarte) > 1:
        descarte = random.choice(descarte)
    else:
        descarte = descarte[0]
        
    cambio = [x for x in options if x not in [primera_opcion,descarte]][0]
    
    result_cambio = 0
    result_azar = 0
    result_no_cambio = 0
    
    if winner == cambio:
        result_cambio = 1
    if winner == primera_opcion:
        result_no_cambio = 1
    if winner == random.choice([primera_opcion,cambio]):
        result_azar = 1
    
    return [result_no_cambio,result_cambio,result_azar]
```

# Run Experiment

```python
import numpy as np
import pandas as pd

n = 10000

experiment = [[0,0,0]]

for trial in range(0,n):
    run = np.add(experiment[trial],monty_iter())
    experiment.append(run)
    
experiment_df = pd.DataFrame(experiment,columns=['result_no_cambio','result_cambio','result_azar'])
```
